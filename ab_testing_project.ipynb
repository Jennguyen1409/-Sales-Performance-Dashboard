```python
import pandas as pd

df = pd.read_csv('ab_test_data.csv')

print("First 5 rows:")
print(df.head())

print("\nDataset shape:")
print(df.shape)
```

```python
print("Group counts:")
print(df['group'].value_counts())

print("\nConversion counts:")
print(df['converted'].value_counts())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
conversion_summary = df.groupby('group')['converted'].agg(['sum', 'count'])
conversion_summary['conversion_rate'] = conversion_summary['sum'] / conversion_summary['count']
conversion_summary['conversion_rate_percent'] = conversion_summary['conversion_rate'] * 100

print(conversion_summary)
```

```python
conversion_summary = df.groupby('group')['converted'].agg(['sum', 'count'])
conversion_summary['conversion_rate'] = conversion_summary['sum'] / conversion_summary['count']
conversion_summary['conversion_rate_percent'] = conversion_summary['conversion_rate'] * 100

print(conversion_summary)

```

```python
# Get counts for each group
control_count = len(df[df['group'] == 'control'])
treatment_count = len(df[df['group'] == 'treatment'])

print(f"Control group users: {control_count}")
print(f"Treatment group users: {treatment_count}")
print(f"Total users: {control_count + treatment_count}")

# Get conversion counts for each group
control_converted = df[df['group'] == 'control']['converted'].sum()
treatment_converted = df[df['group'] == 'treatment']['converted'].sum()

print(f"\nControl group conversions: {control_converted}")
print(f"Treatment group conversions: {treatment_converted}")

# Calculate conversion rates
control_rate = (control_converted / control_count) * 100
treatment_rate = (treatment_converted / treatment_count) * 100

print(f"\nControl group conversion rate: {control_rate:.2f}%")
print(f"Treatment group conversion rate: {treatment_rate:.2f}%")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Prepare the data
conversions = [17723, 17514]  # [control_conversions, treatment_conversions]
users = [147202, 147276]       # [control_users, treatment_users]

# Run z-test
z_stat, p_value = proportions_ztest(conversions, users)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("\n✅ Reject the null hypothesis.")
    print(f"The difference IS statistically significant (p = {p_value:.4f} < 0.05).")
    if 17723/147202 > 17514/147276:
        print("The OLD page performs better than the new page.")
    else:
        print("The NEW page performs better than the old page.")
else:
    print("\n❌ Fail to reject the null hypothesis.")
    print(f"The difference is NOT statistically significant (p = {p_value:.4f} >= 0.05).")
    print("The two pages perform the same.")
```
